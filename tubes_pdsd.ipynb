```python
# Menambahkan column tanggal dan manipulasi tipe data pada
df['tanggal'] = pd.to_datetime(df['dteday'])

# Menambahkan column hari
df['day'] = df['tanggal'].dt.day_name()

# Menambahkan column bulan
df['month'] = df['tanggal'].dt.month_name()

# Menambahkan column tahun
df['year'] = df['tanggal'].dt.year
```

# Visualisasi & EDA

```python
df.head(2)
```

```python
# Membuat visualisasi tren penyewaan sepede pada tahun 2011 bulan januari
df_2011 = df[df['year'] == 2011]
df_2011_jan = df_2011[df_2011['month'] == 'January']

plt.figure(figsize=(12, 6))
plt.plot(
    df_2011_jan['tanggal'],
    df_2011_jan['cnt']
)

plt.title('Tren Penyewaan Sepeda Tahun 2011 Bulan Januari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')


plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%a %d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))

plt.gcf().autofmt_xdate()
plt.show()
```

```python
# Membandingan penyewaan sepeda berdasarkan hari
rental_comparison = df_2011_jan.groupby('day')['cnt'].sum().reset_index()
rental_comparison.sort_values(by='cnt', ascending=False, inplace=True)

plt.figure(figsize=(10, 6))
plt.pie(
    rental_comparison['cnt'],
    labels=rental_comparison['day'],
    autopct='%1.1f%%'
)

plt.title('Perbandingan Penyewaan Sepeda Berdasarkan Hari')
plt.show()
```

```python
# Membuat visualisasi total penyewaan sepeda per hari selama satu bulan
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_2011_jan,
    x='tanggal',
    y='cnt'
)

plt.title('Total Penyewaan Sepeda per Hari (Januari 2011)')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%a %d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
plt.show()
```

```python
# Membuat visualisasi rata rata penyewaan sepeda per hari dalam seminggu
rata_cnt = df_2011_jan.groupby('day')['cnt'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(
    data=rata_cnt,
    x='day',
    y='cnt'
)

plt.xlabel('Hari')
plt.ylabel('Rata-rata Penyewaan')
plt.title('Rata-rata Penyewaan Sepeda per Hari (Januari 2011)')
plt.show()
```

```python
# Membuat visualisasi pengaruh suhu terhadap Penyewaan Sepeda
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='temp',
    y='cnt',
    data=df_2011_jan
)
plt.title('Pengaruh Suhu terhadap Penyewaan Sepeda')
plt.xlabel('Suhu (Normalized)')
plt.ylabel('Total Penyewaan')
plt.show()
```

```python
# Membuat visualisasi pengaruh kelembaban terhadap Penyewaan Sepeda
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='hum',
    y='cnt',
    data=df_2011_jan)
plt.title('Pengaruh Kelembaban terhadap Penyewaan Sepeda')
plt.xlabel('Kelembaban (Normalized)')
plt.ylabel('Total Penyewaan')
plt.show()
```

```python
# Membuat Visualisasi pengaruh kecepatan Angin terhadap Penyewaan Sepeda
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='windspeed',
    y='cnt',
    data=df_2011_jan)
plt.title('Pengaruh Kecepatan Angin terhadap Penyewaan Sepeda')
plt.xlabel('Kecepatan Angin (Normalized)')
plt.ylabel('Total Penyewaan')
plt.show()
```
